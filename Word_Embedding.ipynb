```python
import nltk
import urllib
import bs4 as bs
import re
from gensim.models import Word2Vec
```

```python
source = urllib.request.urlopen("https://en.wikipedia.org/wiki/Sachin_Tendulkar").read()
```

```python
source
```

```python
soup = bs.BeautifulSoup(source,'lxml')
```

```python
soup
```

```python
text = ""
for paragraph in soup.find_all('p'):
  text += paragraph.text
```

```python
text
```

```python
text = re.sub(r"\[[0-9]*\]", ' ', text)
```

```python
text
```

```python
text = text.lower()
text = re.sub(r'[@#\$%\*\(\)\<\>\?\'\",:;]', ' ', text)
```

```python
text
```

```python
text = re.sub(r'\n[a-z]*',' ',text)
text = re.sub(r'/st n t ndu lkr/',' ',text)
text = re.sub(r'\[stin te ulk\]',' ',text)
text = re.sub(r'\s+',' ',text)
```

```python
text
```

```python
nltk.download('punkt')
sentences = nltk.sent_tokenize(text)
```

```python
sentences
```

```python
len(sentences)
```

```python
sentences = [ nltk.word_tokenize(sentence) for sentence in sentences]
```

```python
sentences
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
for i in range(len(sentences)):
    sentences[i] = [word for word in sentences[i] if word not in stopwords.words('english')]
```

```python
nltk.download('stopwords')
stopwords.words('english')
```

```python
sentences
```

```python
model = Word2Vec(sentences,min_count=2)
```

```python
model.wv['sachin']
```
